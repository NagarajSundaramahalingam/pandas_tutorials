# Pandas

What is Pandas?

**Pandas** is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language. [Pandas official page](https://pandas.pydata.org/)

## DataFrame & Series
**DataFrame** is a 2-dimensional labeled data structure with columns of potentially different types. 

**Series** is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.

```
# Importing Pandas

import pandas as pd
```

```
# Creating dataframe by passing a dict object

sample_df = pd.DataFrame({'Name': ['Ram', 'Raj', 'Raju', 'Raja', 'Ravi', 'Ramu'],
     'Tamil': [88, 70, 40, 20, 65, 88],
     'English': [75, 63, 28, 13, 53, 77],
     'Mathematics': [95, 82, 43, 5, 76, 66],
     'Science': [80, 65, 55, 28, 72, 73],
     'SocialScience': [78, 68, 47, 38, 73, 67]
     })
```

```
# Displaying dataframe.

sample_df
```

Let's investigate the sample_df.

sample_df looks like table (row, column) structured data. First column wuithout label is called **Index** (Serial no starts with '0' by default). Pandas creates an Index by default if index is not specified.
