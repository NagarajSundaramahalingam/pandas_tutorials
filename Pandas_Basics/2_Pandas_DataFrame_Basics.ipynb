```
# Importing library

import pandas as pd
import numpy as np
```

```
# Creating dataframe by passing a dict object

sample_df = pd.DataFrame({'Name': ['Ram', 'Raj', 'Raju', 'Raja', 'Ravi', 'Ramu'],
     'Tamil': [88, 70, 40, 20, 65, 88],
     'English': [75, 63, 28, 13, 53, 77],
     'Mathematics': [95, 82, 43, 5, 76, 66],
     'Science': [80, 65, 55, 28, 72, 73],
     'SocialScience': [78, 68, 47, 38, 73, 67]
     })
```

```
sample_df
```

Let's do some basic details about about DataFrame.

```
# index
  
sample_df.index
```

Here Index is a Range Index starts from 0. Pandas by default creates a Range Index (Serial no. from '0') if index is not specified.

```
# columns

sample_df.columns
```

Returns the column names of a DataFrame.

```
# dtypes

sample_df.dtypes
```

The data type of each column.

```
# size

sample_df.size
```

Returns total number of elements. In this case, 6 rows and 6 columns. 6*6


```
# shape

sample_df.shape
```

Returns a tuple representing the row count, column count of the DataFrame.

```
# empty

sample_df.empty
```

If DataFrame is empty, return True, if not return False.

```
# More on empty

df_empty = pd.DataFrame({'A' : []})

df_nan = pd.DataFrame({'A' : [np.nan]})

print(df_empty.empty)
print(df_nan.empty)

df_nan
```

NaN is not considered empty!

```
# head

sample_df.head()
```

Returns first 5 rows by default. We can specify the no. of rows as parameter in head function.

```
# tail

sample_df.tail()
```

Same as head. Reading from bottom to top.

```
# info

sample_df.info()
```

Returns summary of a DataFrame.
